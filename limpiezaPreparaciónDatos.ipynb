El método `.set_index()` es para  identificar cada fila de manera única. Por lo que entiendo, convierte la columna o columnas seleccionadas en un índice y no posición que etiqueta a todas las filas de esa columna con un nombre que facilita la manipulación de los datos.

Aquí usamos `.loc()` para para buscar datos según el nombre de las fila (que sea el índice) o también para seleccionar columnas específicas.

```python
movieTemp.loc[['Coco','Moana','The Avengers']]
```

Vamos a eliminar duplicados y quedarnos con el primero usando `.duplicated()` con el argumento `keep ='first'` y seleccionando el índice con `.index`. Así el código revisa si hay valores repetidos en el índice y se queda con la primera aparición de este. El output del código es una lista de booleans que dan `True` o `False` según si esa fila sea o no un duplicado, menos la primera aparición.

La línea `~` al principio es una negación, que invierte los valores booleanos, y por lo tanto, deja el dataframe únicamente con las columnas que no están repetidas y elimina las demás.

```python
movieTemp = movieTemp[~movieTemp.index.duplicated(keep='first')]
movieTemp.loc[['Coco','Moana','The Avengers']]
```
